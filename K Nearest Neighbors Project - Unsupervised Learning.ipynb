# Classified Dataset - KNN Unsupervised Learning
In this project, we are given a classified datasets and we will try to create a model and train it using unsupervised learning. 

## Import Libraries
**Import pandas,seaborn, and the usual libraries.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Importing the Data
**Read the 'KNN_Project_Data csv file into a dataframe. **

```python
df = pd.read_csv('KNN_Project_Data')
```

**Checking the head of the dataframe.**

```python
df.head() 
```

# Exploratory Data Analysis (EDA)

**Using seaborn on the dataframe to create a pairplot with the hue indicated by the TARGET CLASS column.**

```python
sns.pairplot(df,hue='TARGET CLASS',palette='viridis')
```

# Standardize the Variables

Now we will try to standardize the variables.

**Importing StandardScaler from Scikit learn.**

```python
from sklearn.preprocessing import StandardScaler
```

**Creating a StandardScaler() object called scaler.**

```python
scaler = StandardScaler()
```

**Fitting scaler to the features.**

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

**Using the .transform() method, we will transform the features to a scaled version.**

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

**Converting the scaled features to a dataframe and then checking the head of this dataframe to make sure the scaling worked.**

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

# Train Test Split

**Use train_test_split to split your data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['TARGET CLASS'],
                                                    test_size=0.30)
```

# Using KNN

**Import KNeighborsClassifier from scikit learn.**

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Choosing a K Value
Here, we will try use the elbow method to pick a good K Value for the model!

**We will create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.**

```python
error_rate = []
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

**Now we will create the following plot using the information from the loop.**

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='black', linestyle='dashed', marker='o',
         markerfacecolor='black', markersize=6)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

## Retraining the model with K Value=20 (Lowest Error)

**Retraining our model with the best K-value=20 and evaluating the model with the classification report and the confusion matrix.**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# NOW WITH K=20
knn = KNeighborsClassifier(n_neighbors=20)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=20')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
